# Nested If

```python
x = int(input())
if x>10:
  print("Your number is above then")
if x>20:
  print("and also above 20!")
else:
  print("But not above 20")
```

```python
a=int(input())
if a>10:
  print(">10")
  print("Inside the top if")
  if(a>10):                                                                     # Must follow INDENTATION
    print(">20")
    print("Inside the nested if")
  else:
    print("<=20")
    print("else part")
print("Outside all if")
```

```python
a=int(input())
if a>10:
  print(">10")
  print("Inside the top if")
  if(a>10):
    print(">20")
    print("Inside the nested if")
    if a>30:                                                                    # Must follow INDENTATION
      print(">30")
      print("Inside the else part nested if of nested if")
    else:
      print("<=30")
      print("Inside the else part nif of nif")
  else:
    print("<=20")
    print("else part")
  print("Outside all if")
```
